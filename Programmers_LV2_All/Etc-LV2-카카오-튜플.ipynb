```python
def solution(s):
    r=[]
    s=sorted(s.replace('{','').replace('}}','').split('},'), key=len)
    for i in s:
        for j in list(map(int,i.split(','))):
            if j not in r:
                r.append(j)
    return r
```

```python
print(solution('{{4,2,3},{3},{2,3,4,1},{2,3}}'))
```
