```python
import heapq as hq
def solution(stock, dates, supplies, k):
    s=[]
    r=0
    for a,b in zip(dates, supplies): hq.heappush(s, [a,b])
    for i in range(1, k+1):
        if len(s)>0 and i==s[0][0]:
            p=hq.heappop(s)
            if stock<2:
                stock+=p[1]
                r+=1
        stock-=1
        print(i, stock, s, r)
    return r
```

```python
print(solution(4, [4,10,15], [20,5,10], 30))
```
