```python
def solution(record):
    p = []
    nickname = {}
    for i in record:
        i_=i.split()
        if i_[0] in ['Enter','Change']:
            nickname[i_[1]] = i_[2]
    for i in record:
        i_=i.split()[:2]
        if i_[0] == 'Enter':
            p.append(f'{nickname[i_[1]]}님이 들어왔습니다.')
        if i_[0] == 'Leave':
            p.append(f'{nickname[i_[1]]}님이 나갔습니다.')
    return p
```

```python
print(solution(["Enter uid1234 Muzi", "Enter uid4567 Prodo","Leave uid1234","Enter uid1234 Prodo","Change uid4567 Ryan"]))
```
