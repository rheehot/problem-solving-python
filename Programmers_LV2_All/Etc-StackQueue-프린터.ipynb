```python
def solution(priorities, location):
    p = []
    pr = list(map(list,zip(priorities, range(len(priorities)))))
    while(pr):
        s = pr.pop(0)
        if not pr:
            p.append(s)
            break
        if s[0] >= max(map(lambda x:x[0],pr)):
            p.append(s)
        else:
            pr.append(s)
    return p.index(p[list(map(lambda x:x[1], p)).index(location)])+1
```

```python
print(solution([2, 1, 3, 2], 2))
```
