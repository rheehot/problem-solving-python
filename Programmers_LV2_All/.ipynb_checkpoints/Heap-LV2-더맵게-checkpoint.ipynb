```python
def solution(scoville, K):
    cnt=0
    while scoville:
        mix=scoville.pop(scoville.index(min(scoville)))+scoville.pop(scoville.index(min(scoville)))*2
        scoville.insert(0,mix)
        cnt+=1
        if mix>=K: break
    return cnt if scoville[0]>=K else -1
```

```python
print(solution([1,2,3,9,10,12],7))
```

-> 시간초과, heapq 사용

heapq 모듈은 이진트리 기반의 최소 힙 자료구조<br>
항상 정렬된 상태로 추가, 삭제됨<br>
그러므로 가장 작은 값은 무조건 인덱스 0에 위치<br>
heapq 임포트 후 -> 리스트 선언 후 -><br>
heapq.heappush(리스트이름, 값)<br>
heapq.heappop(리스트이름)

```python
import heapq as h
def solution(scoville, K):
    s = []
    cnt=0
    for i in scoville: h.heappush(s, i)
    while len(s)>1:
        if s[0]>=K: break
        h.heappush(s, h.heappop(s)+h.heappop(s)*2)
        cnt+=1
    return cnt if s[0]>=K else -1
```

```python
print(solution([1,2,3],30))
```
