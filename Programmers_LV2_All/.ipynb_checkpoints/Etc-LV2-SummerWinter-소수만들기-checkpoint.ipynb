```python
from itertools import combinations
def solution(nums):
    s=list(map(sum,list(combinations(nums, 3))))
    n=max(s)
    a = [False,False] + [True]*(n-1)
    primes=[]
    for i in range(2,n+1):
        if a[i]:
            primes.append(i)
            for j in range(2*i, n+1, i):
                a[j] = False
    return len([i for i in s if i in primes])
```

```python
print(solution([1,2,3,4]))
```
