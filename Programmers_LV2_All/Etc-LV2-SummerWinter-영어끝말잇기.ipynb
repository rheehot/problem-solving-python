```python
import math
def solution(n, words):
    u=[]
    for i, w in enumerate(words):
        if i>0 and (u[-1][-1] != w[0] or w in u):
            return [(i+1)%n if (i+1)%n!=0 else n,math.ceil((i+1)/n)]
        u.append(w)
    return [0,0]
```

```python
print(solution(3, ['tank', 'kick', 'know', 'wheel', 'land', 'dream', 'mother', 'robot', 'tank']))
```
