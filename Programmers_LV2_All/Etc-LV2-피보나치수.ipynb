```python
def solution(n):
    def pibo(n):
        if n==0: return 0
        if n==1: return 1
        return pibo(n-1)+pibo(n-2)
    return pibo(n)%1234567
```

```python
print(solution(100000))
```

-> 재귀 초과

```python
def solution(n):
    p=[0,1,1]
    for i in range(3, n+1):
        p.append(p[i-1]+p[i-2])
    return p[n]%1234567
```

```python
print(solution(100000))
```
