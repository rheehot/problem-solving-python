```python
def solution(people, limit):
    people.sort()
    boat = 0
    person = 0
    weight = 0
    while people:
        if weight+people[0]>limit or person==2:
                boat+=1
                weight=0
        weight += people.pop(0)
        person += 1
        print(people, boat, weight)
    return boat
```

```python
print(solution([70, 50, 80, 50], 100))
```
