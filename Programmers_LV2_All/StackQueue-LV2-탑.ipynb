```python
def solution(heights):
    a=[]
    for i in range(len(heights)):
        k = heights.pop()
        for idx, j in enumerate(heights[::-1]):
            if k < j:
                a.append(len(heights)-idx)
                break
        else: a.append(0)
    return a[::-1]
```

```python
print(solution([1,5,3,6,7,6,5]))
```
