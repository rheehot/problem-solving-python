```python
def solution(name_list):
    s = set(name_list)
    for search in s:
        match = 0
        for name in name_list:
            if search in name:
                match += 1
        if match > 1:
            return True
    return False
                

solution(["가을", "우주", "너굴", "김가을"])
```

```python
def sum(base, other):
    for i in other:
        if i not in base:
            base.append(i)
    return base

sum([1,2,3,2,1],[1,3])
```

```python
def complement(base, other):
    for i in other:
        if i in base:
            del base[base.index(i)]
    return base

complement([1,2,3,4],[3,4,5])
```

```python
def intersect(base, other):
    intersect_ = []
    for b in base:
        for o in other:
            if b == o:
                intersect_.append(b)
                break
    return intersect_

intersect([1,2,3],[1,3])
```

```python
def makeset(array):
    set_ = []
    for i in array:
        if i not in set_:
            set_.append(i)
    return set_

makeset([1,2,3,2,1])
```
