```python
def solution(skill, skill_trees):
    cnt = 0
    for t in skill_trees:
        a = [t.index(i) for i in skill if i in t]
        a_ = sorted(a)
        if a == a_ and all(i in t for i in skill[:len(a)]):
            cnt += 1
    return cnt

print(solution("CBD", ["BACDE", "CBADF", "AECB", "BDA"]))
```
