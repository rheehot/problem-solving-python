```python
#participant: 참가자, completion: 완주자
#answer: 완주 못한 참가자 한명
def solution(participant, completion):
    answer = ''
    for i in completion:
        del participant[participant.index(i)]
    answer = participant[0]
    return answer
```

```python
p = ['leo', 'kiki', 'eden']
c = ['eden', 'kiki']
print(solution(p,c))
```

### -> 효율성 테스트 불통과..

```python
#participant: 참가자, completion: 완주자
#answer: 완주 못한 참가자 한명
def solution(participant, completion):
    answer = ''
    participant.sort()
    completion.sort()
    for i in range(len(completion)):
        if participant[i] != completion[i]:
            return participant[i] #중간에 미완주자가 나오는 경우
    return participant[len(participant)-1] #마지막에 있는 경우
```

```python
p = ['leo', 'kiki', 'eden']
c = ['eden', 'kiki']
print(solution(p,c))
```

### 통과
- Collections

```python
import collections


def solution(participant, completion):
    answer = collections.Counter(participant) - collections.Counter(completion)
    return list(answer.keys())[0]
```

```python
p = ['leo', 'kiki', 'eden']
c = ['eden', 'kiki']
print(solution(p,c))
```
